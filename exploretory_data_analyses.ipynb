```python
plt.figure(figsize=(5,6))
plt.scatter(data['Age'],data['Fare'],color="yellow")
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.hist(data['Age'],bins=30,color="red",edgecolor="black")
plt.title("Histogram : Age", fontsize=15)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
